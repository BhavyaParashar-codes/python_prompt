```python
from google.colab import files
uploaded = files.upload()
json_file = list(uploaded.keys())[0]

```

```python
import json
import pandas as pd

# Load the JSON
with open(json_file, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Prepare a list of records
rows = []
chapter_number = data.get("chapter_number", "")
chapter_name = data.get("chapter_name", "")

for topic in data.get("topics", []):
    topic_page_number = topic.get("topic_number", "")
    topic_title = topic.get("topic_title", "")
    for content in topic.get("content", []):
        row = {
            "Chapter Number": chapter_number,
            "Chapter Name": chapter_name,
            "Topic Number": topic_page_number,
            "Topic Title": topic_title,
            "Page No": content.get("page_no", ""),
            "Sequence No": content.get("sequence_no", ""),
            "Type": content.get("type", ""),
            "Internal Name": content.get("internal_name", ""),
            "Actual Content": content.get("actual_content", ""),
        }
        rows.append(row)

# Create DataFrame
df = pd.DataFrame(rows)
```

```python
excel_filename = "hesc113.xlsx"
df.to_excel(excel_filename, index=False)
files.download(excel_filename)
```

```python
!pip install pdfplumber
from google.colab import files
import pdfplumber

#uploaded = files.upload()
#pdf_filename = list(uploaded.keys())[0]
pdf_filename = "hesc113.pdf"

# Extract text page-wise
raw_text = ""
with pdfplumber.open(pdf_filename) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            raw_text += text + "\n\n"

# Save raw text for inspection
with open("hesc113.txt", "w", encoding="utf-8") as f:
    f.write(raw_text)

print("✅ Raw extraction complete. Download below:")
files.download("hesc113.txt")
```
